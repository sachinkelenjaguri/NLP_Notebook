```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs

```

```python

coqa = pd.read_json('http://downloads.cs.stanford.edu/nlp/data/coqa/coqa-train-v1.0.json')
del coqa["version"]
```

```python
cols = ["context","qas"]

# j = 1
comp_list = []
for index, row in coqa.iterrows():
    for i in range(len(row["data"]["questions"])):
        temp_list = []
#         temp_list.append(j)
        temp_list.append(row["data"]["story"])
        temp_list.append( [{'answers': [{'answer_start': row["data"]["answers"][i]['span_start'], 'text': row["data"]["answers"][i]['input_text']}],
                            'id': row["data"]["answers"][i]['turn_id'],
                            'is_impossible': False,
                            'question':  row["data"]["questions"][i]['input_text']}] )
        #temp_list.append(row["data"]["questions"][i]["input_text"])
        
        comp_list.append(temp_list)
#     j += 1
new_df = pd.DataFrame(comp_list, columns=cols) 
```

```python
from sklearn.model_selection import train_test_split

train, testandval = train_test_split(new_df, test_size=0.4)
```

```python
test, val = train_test_split(testandval, test_size=0.5)
```

```python
train1 =[]
for index, row in train.iterrows():
  train1.append({'context':row['context'],'qas':row['qas']})

test1 =[]
for index, row in test.iterrows():
  test1.append({'context':row['context'],'qas':row['qas']})


val1 =[]
for index, row in val.iterrows():
  val1.append({'context':row['context'],'qas':row['qas']})
   
```
